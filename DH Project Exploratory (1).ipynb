```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Arrest_Data_from_2020_to_Present.csv')
```

```python
data
```

```python
data['Descent Code'] = data['Descent Code'].map({'H': 'Hispanic/Latin/Mexican', 
                          'A': 'Other Asian',
                          'B': 'Black',
                          'C' : 'Chinese', 
                          'D' : 'Cambodian',
                          'F' : 'Filipino',
                          'G' : 'Guamanian',
                          'I':'American Indian/Alaskan Native',
                          'J' : 'Japanese',
                          'K' : 'Korean',
                          'L' : 'Laotian',
                          'O' : 'Other',
                          'P' : 'Pacific Islander',
                          'S' : 'Samoan',
                          'U' : 'Hawaiian',
                          'V' : 'Vietnamese',
                          'W' : 'White',
                          'X' : 'Unknown',
                          'Z' : 'Asian Indian'})
```

```python
data['Report Type'] = data['Report Type'].map({
                        'RFC': 'RFC', 
                        'BOOKING': 'BOOKED',
                          })
```
